# Matplotlib

¿Qué es Matplotlib? Matplotlib es una biblioteca para la generación de gráficos a partir de datos contenidos en listas o arrays en el lenguaje de programación Python y su extensión matemática NumPy. Proporciona una API, pylab, diseñada para recordar a la API de MATLAB.


Antes de comenzar con Matplotlib existen algunos aspectos que es importante diferenciar. Al igual que cuando usamos Pandas como `pd` or Numpy como `np`, Matplotlib se suele llamar por:

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

`pyplot` o `plt` es la que se usa mayoritariamente, y es la herramienta que usaremos en este curso.

## Styles

Pyplot tiene una serie de diversos templates o styles, con los que podemos darle diferentes aspectos a nuestros graficos o visualizaciones. Estos aspectos los pueden encontrar en: https://matplotlib.org/3.1.0/gallery/style_sheets/style_sheets_reference.html.

```python
# Vamos hacer uso del style classic
plt.style.use('default')
```

## Como desplegar una figura

Desplegar una figura es distinto en eel caso de que estemos ejecutando un _script_ de Python, o si estamos ejecutando un _notebook_.

**Script python:** si nuestro script `script_python.py` tiene que desplegar una figura, entonces la funcion plt.show() nos ayudará. Esto porque plt.show() buscar por ventanas abiertas en el entorno y despliega una nueva ventana con la figura que deseas mostrar.

Ahora, como nosotros estamos trabajando en un notebook, **no nos tenemos que preocupar por eso**.

```python
# Llamamos a Numpy para hacer uso de arreglos y generar funciones
import numpy as np
```

```python
# Generamos un linespace, en este caso con números de 0 a 8, donde hay 200 puntos intermedios
x = np.linspace(0, 8, 200)
```

```python
# Generamos una función sin(x) a partir de la variable x
sin_x = np.sin(x)
```

```python
plt.plot(x, sin_x)
```

En este caso vamos a usar _Air Passengers_, que registra el número de pasajeeros en una aerolínea en cada mes de un año determinado.

Como vemos, en matplotlib necesitamos entregar las distintas columnas para graficar una función. Ahora vamos a ver que también es posible generar múltiples funciones en el mismo gráfico.

```python
# Funcion exponencial
exp_x = np.exp(x/4)
plt.plot(x, sin_x)
plt.plot(x, exp_x)
```

```python
# Hay que tener mucho cuidado al querer desplegar funciones que no se encuentran cercanas 
# ya que matplotlib intenta ajustar automaticamente
exp_x = np.exp(x/4) + 100
```

```python
plt.plot(x, sin_x)
plt.plot(x, exp_x)
```

Como vemos, es mala idea "mezclar peras con manzanas". Veamos otro ejemplo de qué pasa cuando graficamos en la misma figura observaciones muy distintas.

```python
plt.plot(x, sin_x)
plt.plot(x, exp_x)
plt.plot(AirPassengers["time"], AirPassengers["AirPassengers"])
```

## Markers

Un Marker es la función que grafica los puntos pasados como parametros:

https://matplotlib.org/api/markers_api.html

El parámetro ms define el tamaño del marker.



```python
# En este caso generaremos un arreglo con valores entre 0 y 8, con 20 puntos.
x = np.linspace(0, 8, 20)

plt.plot(x, np.sin(x/16), marker = 'H', markersize=5)
plt.plot(x, np.cos(x), marker = 's', ms = 8);
```

```python
plt.plot(AirPassengers["time"], AirPassengers["AirPassengers"], marker = 'H', markersize=5)
```

## Colores

Las paletas de colores pueden ser encontradas en:

https://matplotlib.org/stable/gallery/color/named_colors.html



```python
x = np.linspace(0, 8, 20)

# Seleccionamos color seagreen
plt.plot(x, np.sin(x), marker = 'o', color= 'seagreen')
# Seleccionamos color red
plt.plot(x, np.cos(x), marker = 'o', color= 'r')
```

Veamos esto mismo sobre el _dataset_ de la aerolínea.

```python
plt.plot(AirPassengers["time"], AirPassengers["AirPassengers"], marker = 'o', color= 'seagreen')
```

## Linestyle

Linestyle es el tipo de linea que se desea desplegar:

https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.lines.Line2D.


```python
x = np.linspace(0, 8, 20)

plt.plot(x, np.sin(x/16), linestyle= '-.', marker = 'o', ms = 5, c= 'b')
plt.plot(x, np.cos(x), ls= '-', marker = "|", ms = 8, c = 'r');
```

Ejemplo con dataset

```python
plt.plot(AirPassengers["time"], AirPassengers["AirPassengers"],linestyle= '-.', marker = 'o', ms = 2, c= 'b')
```

También podemos editar el ancho de las líneas (`linewidth`).

```python
x = np.linspace(0, 8, 20)

plt.plot(x, np.sin(x/16), linewidth = 10 , ls= '--', marker = 'o', ms = 8, c= 'b')
plt.plot(x, np.cos(x), lw= 1, ls= ':', marker = "|", ms = 8, c = 'r')
```

Veamos otro ejemplo con el dataset de la aerolínea.

```python
plt.plot(AirPassengers["time"], AirPassengers["AirPassengers"],linewidth = 8 ,ls= '--', marker = 'o', ms = 8, c= 'b')
```

También podemos agregar transparencia (`alpha`).

```python
x = np.linspace(0, 8, 20)

plt.plot(x, np.sin(x/16), linewidth = 5 ,ls= '--', marker = 'o', ms = 8, c= 'b', alpha = 0.5)
plt.plot(x, np.cos(x), ls= ':', marker = "|" ,ms = 8, c = 'r');
```

Ejemplo con nuestro dataset.

```python
plt.plot(AirPassengers["time"], AirPassengers["AirPassengers"],linewidth = 8 ,ls= '--', marker = 'o', ms = 8, c= 'b',alpha = 0.5)
```

## Figuras, titulos, ejes y legenda

Para poder manipular nuestro gráfico es importante crear un objeto de tipo `Figure`. Este permite hacer modificaciones al espacio del gráfico.

```python
x = np.linspace(0, 8, 20)
# Se debe definir la figura
fig = plt.figure()

# A la figura se le asigna un titulo
fig.suptitle('Esto es un titulo')
# Hacemos los plot
plt.plot(x, np.sin(x/16), linewidth = 5 ,ls= '--', marker = 'o', ms = 8, c= 'b', alpha = 0.5)
plt.plot(x, np.cos(x), ls= ':', marker = "|",ms = 8, c = 'r')
# Definimos los nombres a desplegar de los ejes
plt.xlabel('eje X')
plt.ylabel('eje Y')
```

```python
fig = plt.figure()
# A la figura se le asigna un titulo
fig.suptitle('Esto es un titulo')
# Hacemos los plot
plt.plot(AirPassengers["time"], AirPassengers["AirPassengers"], linewidth = 8 , ls= '--', marker = 'o', ms = 8, c= 'b',alpha = 0.5)
# Definimos los nombres a desplegar de los ejes
plt.xlabel(AirPassengers["time"].name)
plt.ylabel(AirPassengers["AirPassengers"].name)
```

También podemos agregar leyendas (para más información podemos ver https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.legend.html)

```python
x = np.linspace(0, 8, 20)

fig = plt.figure()
# El parametro Label, nos permite dar la descripción de lo que se está desplegando
plt.plot(x, np.sin(x), linewidth = 5 , ls= '--', marker = 'o', ms = 8, c= 'b', alpha = 0.5, label='Sin(x)')
plt.plot(x, np.cos(x), ls= ':', marker = "|", ms = 8, c = 'r', label='cos(x)')
plt.xlabel('X ')
plt.ylabel('Y ')
# La función legend despliega los labels de cada plot, como parametro se le puede entregar la ubicación, por default intenta la mejor posición
plt.legend(loc = 1)
fig.suptitle('Esto es un titulo')
```

Ejemplo con nuestro _dataset_.

```python
fig = plt.figure()
# A la figura se le asigna un titulo
fig.suptitle('Esto es un titulo')
# Hacemos los plot (ahora sin linewidth)
plt.plot(AirPassengers["time"], AirPassengers["AirPassengers"], ls= '--', marker = 'o', ms = 8, c= 'b',alpha = 0.5, label="cantidad de pasajeros")
# Definimos los nombres a desplegar de los ejes
plt.xlabel(AirPassengers["time"].name)
plt.ylabel(AirPassengers["AirPassengers"].name)
# La función legend despliega los labels de cada plot, como parametro se le puede entregar la ubicación, por default intenta la mejor posición
plt.legend(loc = 'best')
```


Vamos ahora cómo editar el **tamaño de una figura**.

```python
x = np.linspace(0, 8, 20)
# Al generar una figura se le puede modificar su tamaño con el parametro figsize
fig = plt.figure(figsize=(18,9))
# El parametro label, nos permite dar la descripción de lo que se está desplegando
plt.plot(x, np.sin(x),linewidth = 5 , ls= '--', marker = 'o', ms = 8, c= 'b', alpha = 0.5, label='Sin(x)')
plt.plot(x, np.cos(x),ls= ':', marker = "|",ms = 8, c = 'r', label='cos(x)')
plt.xlabel('X ')
plt.ylabel('Y ')
# La función legend despliega los labels de cada plot, como parametro se le puede entregar la ubicación, por default intenta la mejor posición
plt.legend(loc = 1)
fig.suptitle('Esto es un titulo');
```

## Multiplot

El uso más común es hacer uso de **subplots** con la función `plt.subplot()`.

```python
fig = plt.figure(figsize=(18,9))
plt.subplot(3, 4, 1)
plt.subplot(3, 4, 2)
plt.subplot(3, 4, 3)
plt.subplot(3, 4, 4)
plt.subplot(3, 4, 5)
plt.subplot(3, 4, 6)
plt.subplot(3, 4, 7)
plt.subplot(3, 4, 8)
plt.subplot(3, 4, 9);
plt.subplot(3, 4, 10);
plt.subplot(3, 4, 11);
```

```python
fig = plt.figure(figsize=(18,9))
for i in range(1, 10):
    plt.subplot(3, 3, i)
```

La funcion `fig.subplot_adjust` sirve para ajustar los espacios entre los graficos.

```python
fig = plt.figure(figsize=(10,5))
fig.subplots_adjust(hspace=0.8, wspace=0.4)

for i in range(1, 10):
    plt.subplot(3, 3, i)
```

La funcion `text()` nos permite ingresar un texto en el grafico.

```python
x = np.linspace(0,10,200)
fig = plt.figure(figsize=(10,5))
fig.subplots_adjust(hspace=.8, wspace=0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str(i), fontsize=18, ha='center', transform=ax.transAxes)
    ax.plot(x,np.sin(x*i))
```

A modo de ejemplo, usaremos AirPassangers pero haremos un zoom sobre el eje del tiempo para mostrar distintos intervalos de tiempo.

```python
AirPassengers["zoom"] = pd.cut(AirPassengers["time"], bins=6, labels=[1, 2, 3, 4, 5, 6])
AirPassengers.head()
```

```python
AirPassengers.groupby('zoom').zoom.count()
```

```python
fig = plt.figure(figsize=(18,9))
fig.subplots_adjust(hspace=.8, wspace=0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str(i), fontsize=18, ha='center',transform=ax.transAxes)
    df_aux = AirPassengers[AirPassengers["zoom"] == i]
    
    # Graficamos
    ax.plot(df_aux["time"], df_aux["AirPassengers"], 
            linewidth = 1 ,ls= '--', marker = 'o', ms = 8,
            c= 'b', alpha = 0.5, label='Air passangers')
```

## Más ejemplos


```python
# Agregar otro plot a un subgráfico
# En este caso es el 5
x = np.linspace(0,10,200)
fig = plt.figure(figsize=(10,5))
fig.subplots_adjust(hspace=.8, wspace=0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str(i), fontsize=18, ha='center',transform=ax.transAxes)
    ax.plot(x,np.sin(x*i))
    if i == 5:
      ax.plot(x,np.cos(x*i/8), c='r')
```

```python
# Títulos
x = np.linspace(0,10,200)
fig = plt.figure(figsize=(18,9))
fig.subplots_adjust(hspace=.8, wspace=0.4)

# Recorrer hasta el ax 5
for i in range(1, 6):
    ax = fig.add_subplot(2, 3, i, title = 'subplot title ' + str(i))
    ax.plot(x,np.sin(x*i))
    ax.set_xlabel('X ' + str(i))
    ax.set_ylabel('Y ' + str(i))
    if i == 5:
      ax.plot(x,np.cos(x*i/8), c='r')


# Generar ax 6 de manera independiente
ax = fig.add_subplot(2, 3, 6)
ax.plot(x, np.sin(x/16),linewidth = 1 ,ls= '--', marker = 'o', ms = 0.5, c= 'b', alpha = 0.5,label='Sin(x) / 16')
ax.plot(x, np.cos(x),ls= ':', marker = "|",ms = 1, c = 'r',label='cos(x)')
ax.set_xlabel('X ')
ax.set_ylabel('Y ')
#titulo del subplot
ax.set_title('Esto es un titulo de un subplot')
ax.legend(loc = 'lower right')

#titulo de la figura
fig.suptitle('Esto es un titulo de la figura',size = '30');
```

```python
# Ahora con el ejemplo de la aerolínea
x = np.linspace(0,10,200)
fig = plt.figure(figsize=(18,9))
fig.subplots_adjust(hspace=.8, wspace=0.4)

# Recorrer hasta el ax 5
for i in range(1, 6):
    ax = fig.add_subplot(2, 3, i, title = 'subplot title ' + str(i))
    df_aux = AirPassengers[AirPassengers["zoom"]==i]
    ax.plot(df_aux["time"], df_aux["AirPassengers"],linewidth = 1 ,ls= '--', marker = 'o', ms = 8, c= 'b', alpha = 0.5,label='Air passangers')
    ax.set_xlabel('Zoom ' + str(i))
    ax.set_ylabel('Cantidad de Pasajeros ')

# Generar ax 6 de manera independiente
ax = fig.add_subplot(2, 3, 6)
df_aux = AirPassengers[AirPassengers["zoom"]==6]
ax.plot(df_aux["time"], df_aux["AirPassengers"],linewidth = 1 ,ls= '--', marker = 'o', ms = 8, c= 'r', alpha = 0.5,label='lastest Air passangers data info')
ax.set_xlabel('last data information')
ax.set_ylabel('Cantidad de Pasajeros ')
# Tíitulo del subplot
ax.set_title('Esto es un titulo de un subplot')
ax.legend(loc = 'lower right')

# Título de la figura
fig.suptitle('Esto es un titulo de la figura',size = '30');
```

```python
with plt.xkcd():
    # Todos los gráficos hechos acá vienen con estilo XKCD
    plt.plot(AirPassengers["time"], AirPassengers["AirPassengers"],linestyle= '-.', marker = 'o', ms = 2, c= 'b')
    plt.xlabel("Time")
    plt.ylabel("Passengers")
    plt.title("Passengers through the years")
```
